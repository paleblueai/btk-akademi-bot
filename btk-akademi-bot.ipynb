```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd

tcno=
sifre=""

class BTK():
    def  __init__ (self,tcno,sifre):
        self.tcno=tcno
        self.sifre=sifre
        self.tarayici=webdriver.Chrome()
        self.tarayici.get("https://www.btkakademi.gov.tr")
        self.tarayici.maximize_window()
        time.sleep(2)
        
    def Login(self):
        
        self.tarayici.find_element_by_xpath("//*[@id='__next']/div/div[1]/div[3]/div/div/div[4]/a/button").click()
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div/form/div/button[1]").click()
        time.sleep(2)
        tcnoInput=self.tarayici.find_element_by_xpath("//*[@id='tridField']")
        tcnoInput.send_keys(self.tcno)
        
        sifreInput=self.tarayici.find_element_by_xpath("//*[@id='egpField']")
        sifreInput.send_keys(self.sifre)
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='loginForm']/div[2]/input[4]").click()
    
    def Profile(self):
        time.sleep(2)
        #profile dropdown
        self.tarayici.find_element_by_xpath("//*[@id='__next']/div/div[1]/div[3]/div/div/div[5]/div/div/a[1]").click()
        time.sleep(1)
        #profile dropdown
        self.tarayici.find_element_by_xpath("//*[@id='dropdown_profile']/ul/a[1]").click()
    
    def egitimlerim(self):
        #ilk 5 eğitimi göstermektedir
        time.sleep(2)
        #eğitimlerim
        self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div/div[1]/div[2]/div[1]/a").click()
        time.sleep(1)
        main=self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[4]/div/div[2]/div[2]/a")
        print(f"Birinci Ders:{main.text}")
        print("***************************")
        
        main1=self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[5]/div/div[2]/div[2]/a/p")
        print(f"İkinci Ders:{main1.text}")
        print("***************************")
        
        
        main3=self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[6]/div/div[2]/div[2]/a/p")
        print(f"Üçüncü Ders:{main3.text}")
        print("***************************")
        
        main4=self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[7]/div/div[2]/div[2]/a/p")
        print(f"Dördüncü Ders:{main4.text}")
        print("***************************")
        
        main5=self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[8]/div/div[2]/div[2]/a/p")
        print(f"Beşinci Ders:{main5.text}")
        print("***************************")

    def enSonSertifika(self):
        time.sleep(1)
        self.tarayici.get("https://www.btkakademi.gov.tr/portal")
        #profile dropdown
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='__next']/div/div[1]/div[3]/div/div/div[5]/div/div/a[1]").click()
        time.sleep(1)
        #profile dropdown
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='dropdown_profile']/ul/a[1]").click()
        #eğitimlerim
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div/div[1]/div[2]/div[1]/a").click()
        #SERTİFİKALARIM
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='cert_btn']").click()
        time.sleep(1)
        sertifikaAdi=self.tarayici.find_element_by_xpath("//*[@id='cert_section']/div/div[2]/div/div[1]/table/tr[1]/td[2]").text
        egitimAdi=self.tarayici.find_element_by_xpath("//*[@id='cert_section']/div/div[2]/div/div[1]/table/tr[1]/td[3]").text
        
        
        print(f"En Son Aldığın Sertifika:\n Sertifika Adı:{sertifikaAdi} \n Eğitim Adı:{egitimAdi}")
        print("***************************")
        
    def ilkBesEgitimDetay(self):
        time.sleep(1)
        self.tarayici.get("https://www.btkakademi.gov.tr/portal")
        #profile dropdown
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='__next']/div/div[1]/div[3]/div/div/div[5]/div/div/a[1]").click()
        time.sleep(1)
        #profile dropdown
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='dropdown_profile']/ul/a[1]").click()
        #eğitimlerim
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div/div[1]/div[2]/div[1]/a").click()
        
        #BİRİNCİEĞİTİM
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[4]/div/div[2]/div[2]/a").click()
        time.sleep(1)
        
        egitimBasligi1=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[2]/p")
        egitimBasligi=egitimBasligi1.text
        
        begenmeSayisi1=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[1]/span")
        begenmeSayisi=begenmeSayisi1.text
        
        begenmemeSayisi1=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[2]/span")
        begenmemeSayisi=begenmemeSayisi1.text
        
        katilimcisayisi1=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/p/span[2]")
        katilimcisayisi=katilimcisayisi1.text
        
        katilimDurumu1=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[1]/a")
        katilimDurumu=katilimDurumu1.text
        
        tamamlamaOrani1=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[5]/a")
        tamamlamaOrani=tamamlamaOrani1.text
        
        print(f"Birinci Eğitim Detayları:\nEğitim Başlığı:{egitimBasligi}\nBeğenme Sayısı:{begenmeSayisi}")
        print(f"Beğenmeme Sayısı:{begenmemeSayisi}\nKatılımcı Sayısı:{katilimcisayisi}\nKatılım Durumu:{katilimDurumu}\nTamamlama Oranı:{tamamlamaOrani}")
        print("***************************")
        print("***************************")
        
        #İKİNCİEĞİTİM
        self.tarayici.get("https://www.btkakademi.gov.tr/portal/courses")
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[5]/div/div[2]/div[2]/a").click()
        time.sleep(1)
        
        egitimBasligi2=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[2]/p")
        egitimBasligi3=egitimBasligi2.text
        
        begenmeSayisi2=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[1]/span")
        begenmeSayisi3=begenmeSayisi2.text
        
        begenmemeSayisi2=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[2]/span")
        begenmemeSayisi3=begenmemeSayisi2.text
        
        katilimcisayisi2=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/p/span[2]")
        katilimcisayisi3=katilimcisayisi2.text
        
        katilimDurumu2=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[1]/a")
        katilimDurumu3=katilimDurumu2.text
        
        tamamlamaOrani2=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[5]/a")
        tamamlamaOrani3=tamamlamaOrani2.text
        
        
        print(f"İkinci Eğitim Detayları:\nEğitim Başlığı:{egitimBasligi3}\nBeğenme Sayısı:{begenmeSayisi3}")
        print(f"Beğenmeme Sayısı:{begenmemeSayisi3}\nKatılımcı Sayısı:{katilimcisayisi3}\nKatılım Durumu:{katilimDurumu3}\nTamamlama Oranı:{tamamlamaOrani3}")
        print("***************************")
        print("***************************")
        
        
        
       
        #ÜÇÜNCÜEĞİTİM
        self.tarayici.get("https://www.btkakademi.gov.tr/portal/courses")
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[6]/div/div[2]/div[2]/a").click()
        time.sleep(1)
        egitimBasligi4=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[2]/p")
        egitimBasligi5=egitimBasligi4.text
        
        begenmeSayisi4=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[1]/span")
        begenmeSayisi5=begenmeSayisi4.text
        
        begenmemeSayisi4=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[2]/span")
        begenmemeSayisi5=begenmemeSayisi4.text
        
        katilimcisayisi4=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/p/span[2]")
        katilimcisayisi5=katilimcisayisi4.text
        
        katilimDurumu4=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[1]/a")
        katilimDurumu5=katilimDurumu4.text
        
        tamamlamaOrani4=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[5]/a")
        tamamlamaOrani5=tamamlamaOrani4.text
        
        
        print(f"Üçüncü Eğitim Detayları:\nEğitim Başlığı:{egitimBasligi5}\nBeğenme Sayısı:{begenmeSayisi5}")
        print(f"Beğenmeme Sayısı:{begenmemeSayisi5}\nKatılımcı Sayısı:{katilimcisayisi5}\nKatılım Durumu:{katilimDurumu5}\nTamamlama Oranı:{tamamlamaOrani5}")
        print("***************************")
        print("***************************")
        
        
        #DÖRDÜNCÜEĞİTİM
        self.tarayici.get("https://www.btkakademi.gov.tr/portal/courses")
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[7]/div/div[2]/div[2]/a").click()
        time.sleep(1)
        egitimBasligi6=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[2]/p")
        egitimBasligi7=egitimBasligi6.text
        
        begenmeSayisi6=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[1]/span")
        begenmeSayisi7=begenmeSayisi6.text
        
        begenmemeSayisi6=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[2]/span")
        begenmemeSayisi7=begenmemeSayisi6.text
        
        katilimcisayisi6=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/p/span[2]")
        katilimcisayisi7=katilimcisayisi6.text
        
        katilimDurumu6=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[1]/a")
        katilimDurumu7=katilimDurumu6.text
        
        tamamlamaOrani6=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[5]/a")
        tamamlamaOrani7=tamamlamaOrani6.text
        
        
        print(f"Dördüncü Eğitim Detayları:\nEğitim Başlığı:{egitimBasligi7}\nBeğenme Sayısı:{begenmeSayisi7}")
        print(f"Beğenmeme Sayısı:{begenmemeSayisi7}\nKatılımcı Sayısı:{katilimcisayisi7}\nKatılım Durumu:{katilimDurumu7}\nTamamlama Oranı:{tamamlamaOrani7}")
        print("***************************")
        print("***************************")
        
        
        #BEŞİNCİEĞİTİM
        self.tarayici.get("https://www.btkakademi.gov.tr/portal/courses")
        time.sleep(1)
        self.tarayici.find_element_by_xpath("//*[@id='main_section']/div[8]/div/div[2]/div[2]/a").click()
        time.sleep(1)
        
        egitimBasligi8=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[2]/p")
        egitimBasligi9=egitimBasligi8.text
        
        begenmeSayisi8=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[1]/span")
        begenmeSayisi9=begenmeSayisi8.text
        
        begenmemeSayisi8=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/div/span[2]/span")
        begenmemeSayisi9=begenmemeSayisi8.text
        
        katilimcisayisi8=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[1]/div/div[1]/div[2]/div[3]/p/span[2]")
        katilimcisayisi9=katilimcisayisi8.text
        
        katilimDurumu8=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[1]/a")
        katilimDurumu9=katilimDurumu8.text
        
        tamamlamaOrani8=self.tarayici.find_element_by_xpath("//*[@id='container-toggle']/div/div/div[1]/div[2]/div[1]/div[1]/p[5]/a")
        tamamlamaOrani9=tamamlamaOrani8.text
        
        
        print(f"Beşinci Eğitim Detayları:\nEğitim Başlığı:{egitimBasligi9}\nBeğenme Sayısı:{begenmeSayisi9}")
        print(f"Beğenmeme Sayısı:{begenmemeSayisi9}\nKatılımcı Sayısı:{katilimcisayisi9}\nKatılım Durumu:{katilimDurumu9}\nTamamlama Oranı:{tamamlamaOrani9}")
        print("***************************")
        print("***************************")
        
    
        data={"Birinci Eğitim Detayları":[egitimBasligi,begenmeSayisi,begenmemeSayisi,katilimcisayisi,katilimDurumu],
              "İkinci Eğitim Detayları":[egitimBasligi3,begenmeSayisi3,begenmemeSayisi3,katilimcisayisi3,katilimDurumu3],
              "Üçüncü Eğitim Detayları":[egitimBasligi5,begenmeSayisi5,begenmemeSayisi5,katilimcisayisi5,katilimDurumu5],
              "Dördüncü Eğitim Detayları":[egitimBasligi7,begenmeSayisi7,begenmemeSayisi7,katilimcisayisi7,katilimDurumu7],
              "Beşinci Eğitim Detayları":[egitimBasligi9,begenmeSayisi9,begenmemeSayisi9,katilimcisayisi9,katilimDurumu9]}
        indexler=["Eğitim Baslıgı","Begenme Sayisi","Begenmeme Sayisi","Katilimci Sayisi","Katilim Durumu"]
        frame=pd.DataFrame(data,index=indexler)
        return frame
        
    
        
        
    
    
        
btk=BTK(tcno, sifre)
btk.Login()
btk.Profile()

btk.egitimlerim()
btk.enSonSertifika()
btk.ilkBesEgitimDetay()


```
